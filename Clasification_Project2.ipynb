```python
max_number_of_nans = 37000
df = df.loc[:, (df.isnull().sum(axis=0) <= max_number_of_nans)]
```

```python
df.shape
```

```python
cat_columns = df.select_dtypes(['category']).columns
df[cat_columns] = df[cat_columns].apply(lambda x: x.cat.codes)
```

```python
df.head()
```

```python
target = 'UPSELLING'
X = df.drop(target,axis=1)
y = df.loc[:,target]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =  train_test_split(X,y,test_size=0.095,random_state=578)
X_train.shape, X_test.shape,y_train.shape
```

```python
X_train,X_train_dev, y_train,y_train_dev = train_test_split(X_train,y_train,test_size=0.05,random_state=578)
X_train.shape,X_train_dev.shape, y_train.shape,y_train_dev.shape
```

```python
from sklearn.compose import make_column_selector
num_col = make_column_selector(dtype_exclude=object)
cat_col = make_column_selector(dtype_include=object)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.pipeline import make_pipeline
```

```python
imp_mean = SimpleImputer(strategy='mean')
imp_median = SimpleImputer(strategy='median')
imp_cat = SimpleImputer(strategy='most_frequent')
onehot = OneHotEncoder(handle_unknown='ignore')
ordencode = OrdinalEncoder(handle_unknown='use_encoded_value',unknown_value=-1)
```

```python
col_transform_mean = make_column_transformer(
    (make_pipeline(imp_mean),num_col),
    (make_pipeline(imp_cat,onehot),cat_col),
    remainder='passthrough'
)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
```

```python
pipe_mean = make_pipeline(col_transform_mean,log_model)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
K_fold = KFold(n_splits=5,shuffle=True,random_state=985)
cross_val_score(pipe_mean,X_train_dev,y_train_dev,cv=K_fold).mean()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier()
pipe_mean_knn = make_pipeline(col_transform_mean,knn_model)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
K_fold = KFold(n_splits=5,shuffle=True,random_state=985)
cross_val_score(pipe_mean_knn,X_train_dev,y_train_dev,cv=K_fold).mean()
```

```python
from sklearn.model_selection import GridSearchCV
params = {
          'kneighborsclassifier__n_neighbors': range(2,15),
          'kneighborsclassifier__p': [1,2,3]
          }
grdcv = GridSearchCV(pipe_mean_knn,param_grid=params,cv=K_fold,n_jobs=-1)
```

```python
from sklearn.model_selection import RandomizedSearchCV
rmdcv = RandomizedSearchCV(pipe_mean_knn,params,random_state=895,n_jobs=-1,cv=K_fold)
```

```python
rmdcv.fit(X_train_dev,y_train_dev)
```
