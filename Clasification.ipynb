```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv("https://raw.githubusercontent.com/asarfraaz/PyPractice/master/ml_data/exam_result.csv")
df.head()
```

```python
df.Result = df.Result.apply(lambda x:1 if x=='Yes' else 0)
```

```python
df
```

```python
df.shape
```

```python
X = df.drop('Result',axis=1)
y = df.Result
```

```python
X.shape,y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
Scaled_values = scaler.fit_transform(X)
```

```python
X_scaled = pd.DataFrame(Scaled_values,columns=X.columns.to_list())
X_scaled.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=3)
```

```python
model.fit(X_scaled,y)
```

```python
std1 = [[75,10]]
```

```python
std1_scale = scaler.transform(std1)
```

```python
model.predict(std1_scale)
```

```python
model.predict_proba(std1_scale)
```

```python
std2 = [[68,9]]
```

```python
std2_scale = scaler.transform(std2)
```

```python
model.predict(std2_scale)
```

```python
model.predict_proba(std2_scale)
```
