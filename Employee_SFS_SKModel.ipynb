```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('https://raw.githubusercontent.com/Manju410/MLPractice/main/data/Employee_Cleanup.csv')
df.head()
```

```
df.shape
```

```
df.columns
```

```
X = df.drop(['SALARY'],axis=1)
y= df.SALARY
```

```
X.shape,y.shape
```

```
from sklearn.model_selection import train_test_split
```

```
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=1234)
```

```
from sklearn.preprocessing import MinMaxScaler
```

```
mmscaler = MinMaxScaler()
```

```
X_train.head()
```

```
continuoscol = ['DEPARTMENT_ID','DEPARTMENT_ID_1','LOCATION_ID','LOCATION_ID_1','REGION_ID','REGION_ID_1']
```

```
scaler = mmscaler.fit(X_train[continuoscol])
```

```
scaler.data_max_
```

```
scaler.data_min_
```

```
scaled_vals = scaler.transform(X_train[continuoscol])
```

```
scaled_vals1 = scaler.transform(X_test[continuoscol])
```

```
X_test_scaled = pd.DataFrame(scaled_vals1,columns=continuoscol,index=X_test.index)
X_test_scaled.head()
```

```
X_train_scaled = pd.DataFrame(scaled_vals,columns=continuoscol,index=X_train.index)
X_train_scaled.head()
```

```
X_train.drop(columns=continuoscol,axis=1,inplace=True)
```

```

X_train.insert(0,X_train_scaled.columns[0],X_train_scaled[X_train_scaled.columns[0]])
X_train.insert(1,X_train_scaled.columns[1],X_train_scaled[X_train_scaled.columns[1]])
X_train.insert(2,X_train_scaled.columns[2],X_train_scaled[X_train_scaled.columns[2]])
X_train.insert(3,X_train_scaled.columns[3],X_train_scaled[X_train_scaled.columns[3]])
X_train.insert(4,X_train_scaled.columns[4],X_train_scaled[X_train_scaled.columns[4]])
X_train.insert(5,X_train_scaled.columns[5],X_train_scaled[X_train_scaled.columns[5]])
```

```

X_train.head()
```

```
X_test.drop(columns=continuoscol,axis=1,inplace=True)
X_test.insert(0,X_test_scaled.columns[0],X_test_scaled[X_test_scaled.columns[0]])
X_test.insert(1,X_test_scaled.columns[1],X_test_scaled[X_test_scaled.columns[1]])
X_test.insert(2,X_test_scaled.columns[2],X_test_scaled[X_test_scaled.columns[2]])
X_test.insert(3,X_test_scaled.columns[3],X_test_scaled[X_test_scaled.columns[3]])
X_test.insert(4,X_test_scaled.columns[4],X_test_scaled[X_test_scaled.columns[4]])
X_test.insert(5,X_test_scaled.columns[5],X_test_scaled[X_test_scaled.columns[5]])
```

```
X_test.head()
```

```
from sklearn.feature_selection import SequentialFeatureSelector
```

```
from sklearn.linear_model import LinearRegression
```

```
linmodel = LinearRegression()
```

```
sfs = SequentialFeatureSelector(linmodel, n_features_to_select=13,cv=10,direction='forward',scoring='neg_mean_squared_error')

sfs1 = sfs.fit(X_train, y_train)
```

```
sfs1.support_
```

```
best_features = [name for name,sup in zip(X_train.columns,sfs1.support_) if sup== True]
best_features
```

```
new_dfXtrain = X_train[best_features]
new_dfXtrain.head()
```

```
new_dfXtest = X_test[best_features]
new_dfXtest
```

```
new_dfXtrain.shape,new_dfXtest.shape
```

```
linmodel.fit(new_dfXtrain,y_train)
```

```
y_pred = linmodel.predict(new_dfXtest)
```

```
df_coeff = pd.DataFrame({'columns':new_dfXtrain.columns,'coeff':linmodel.coef_})
df_coeff
```

```
residuals = pd.DataFrame({'pred':y_pred,'actul':y_test})
residuals.head()
```

```
from sklearn.metrics import r2_score
```

```
r2_score(y_pred,y_test)
```
