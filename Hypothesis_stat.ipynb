```python
import numpy as np
import scipy.stats
```

- Avg marks of students in maths is 85%
- Teacher wants to check this has changed after online classes

- H0: X' == mu 
- H1: X' != mu

```python
scores = [74.5,87.1,80.4,79.8,81.2,75.3,77.5,83.4,79.1,88.2]
```

- since we only have population mean so we are going with T test
- this is a two tail t-test
- sample size is 10

```python
n = len(scores)
n
```

```python
X_bar = np.array(scores).mean()
X_bar
```

```python
X_arr = np.array(scores)
```

```python
X_sd = X_arr.std(ddof=1)
X_sd
```

```python
T_stat = (X_bar - 85)/(X_sd/(np.sqrt(n)))
T_stat
```

```python
T_Crit = 2.262
```

- T_stat > Tcrit so we have to reject H0
- Avg marks of students has been changed after the online classes

```python
T_stat1 =  scipy.stats.ttest_1samp(scores,85)
T_stat1
```

- p-value = 0.014 this is less than 0.025 so we have to reject H0
