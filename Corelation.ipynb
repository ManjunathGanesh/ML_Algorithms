```python
import pandas as pd
```

```python
DFcar = pd.DataFrame({ 'Age' : [2, 1.5, 3.2, 5.4, 2.7], 'Mileage' : [12.4, 13.2, 12.8, 9.3,12.7]})                                                     
```

```python
DFcar.head()
```

```python
meanage , meanmilg = DFcar.mean()
```

```python
meanage
```

```python
meanmilg
```

```python
DFcar['v1'] = DFcar['Age']-meanage
```

```python
DFcar.head()
```

```python
DFcar['v2'] = DFcar['Mileage']-meanmilg
```

```python
DFcar.head()
```

```python
DFcar['prod'] = DFcar['v1']*DFcar['v2']
```

```python
DFcar.head()
```

```python
Sumprod = DFcar['prod'].sum()
```

```python
row,col = DFcar.shape
```

```python
row
```

```python
cov = Sumprod/(row-1)
```

```python
cov
```

```python
stdage = DFcar['Age'].std(ddof=1)
```

```python
stdage, stdmil = DFcar[['Age','Mileage']].std(ddof=1)
```

```python
stdage
```

```python
stdmil
```

```python
R = cov/(stdage*stdmil)
```

```python
R
```
