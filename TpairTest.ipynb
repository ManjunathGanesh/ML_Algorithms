- H0 : There is no effect of treatment (Xpre == Xpost)
- H1 : There is change in the treatment (Xpre != Xpost)
- It is Two tail test


```python
DF["Difference"]= DF['Offline']-DF['Online']
```

```python
DF.head()
```

```python
N = len(DF.Online)
```

```python
DiffMean = DF['Difference'].mean()
DiffMean
```

```python
DiffStd = DF['Difference'].std(ddof=1)
DiffStd
```

```python
Tstat = (DiffMean/(DiffStd/(N**0.5)))
Tstat
```

- Tcritical from Table for 95% significance level
- alpha value = 0.05

```python
DOF = N-1
DOF
```

- Tcritical from table = 2.074
- Tstat > Tcritical so we Reject H0
- so Treatment has some effect after online calss.

-P value = 0.03 less than 0.05 so we reject H0

-There were 48 handouts that were sold We want to know if all 4 models of handsets are equally liked Expected sales per model = 48 / 4 = 12

- H0 : All models are equally liked
- H1 : All models are not equally liked

```python
Expected = [12,12,12,12]
Observed = [11,14,15,8]
```

```python
scipy.stats.chisquare(Observed,Expected)
```
