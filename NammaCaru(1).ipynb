```python
#x=x.replace('*','')
Cardf.loc[:,'workclass']=Cardf['workclass'].str.replace('?','No-Work')
Cardf.head()
```

```python
Cardf.loc[:,'occupation']=Cardf['occupation'].str.replace('?','No-Occupation')
Cardf.head()
```

Workclass and occupation column contain non valuable data ie(?) - so replaced those data with not working and no occupation

```python
Cardf['native-country'].unique()
```

```python
Cardf.drop(Cardf[Cardf['native-country']==' ?'].index, inplace=True)
```

Native-Country column has '?' values: droping rows contain '?' because persons dont know there country doesnot give meaningful information.

```python
#Cardf['Ageinfo']=[x for x in Cardf['age'] if x<=20: return 'Teens' elif: x>=70 return 'SuperSeniors' elif: x>=60 return 'Seniors']
```

```python
def CalcAge(x):
    if x<=20: return 'Teens'
    elif x>=70: return 'SuperSeniors'
    elif x>=60: return 'Seniors'
    else: return 'adults'
    
```

```python
Cardf['Ageinfo']=[CalcAge(x) for x in Cardf['age']]
```
