# Best value of n_neighbors from 2-30 and p value(distance) from 1 to 3

```python
from sklearn.model_selection import cross_val_score
```

```python
mean_scores = {}
for pval in range(1,4):
  for n_fold in range(2,30):
      classifier = KNeighborsClassifier(p=pval,n_neighbors=n_fold)
      score = cross_val_score(classifier,X,y,scoring='r2',cv=10)
      mean_scores[str(pval)+":"+str(n_fold)]= score.mean()
      #mean_scores.append(str(pval)+":"+str(n_fold)+":"+str(score.mean()))
```
