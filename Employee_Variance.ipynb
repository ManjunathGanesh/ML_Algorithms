```
df.columns
```

```
dominate_cat =[]
for element in catcols:
  most_freq = df[element].value_counts().mul(100).div(len(df))[0]
  if most_freq>50:
    dominate_cat.append(element)

df = df.drop(dominate_cat,axis=1)
```

```
uniqucount = []
for element in catcols:
  if element in df and df[element].nunique()>10:
    uniqucount.append(element)

df = df.drop(uniqucount,axis=1)
```

```
df.columns
```

```
continuecol
```

```
for feature in continuecol:
    if feature in df and df[feature].nunique() > 10:
        plt.figure()
        plt.boxplot(df[feature], whis=2)
        plt.title(f'{feature}', fontsize=20)
        sns.despine()
```

```
print(f'\033[1mNumber of rows before removing the outliers:\033[0m {df.shape[0]}')

for feature in continuecol:
    if feature in df:
        Q1 = df[feature].quantile(0.25)
        Q3 = df[feature].quantile(0.75)
        IQR = Q3 - Q1
        df=df.drop(df[df[feature] > (Q3 + 2 * IQR)].index)
df = df.reset_index(drop=True)

print(f'\033[1mNumber of rows after removing the outliers:\033[0m  {df.shape[0]}')
```

```
df.head()
```

```
catcols1 = df.dtypes[df.dtypes =='object'].index.tolist()
catcols1
```

```

def onehotfun(df,catcol):
  onehotdummies = []
  onehotdummies.append(df)
  for cat in catcol:
    catname = pd.get_dummies(df[cat],prefix=cat)
    onehotdummies.append(catname)

  dffinal = pd.concat(onehotdummies,axis=1)
  dffinal.drop(columns=catcol, axis=1, inplace=True)
  return dffinal
```

```
finaldf = onehotfun(df,catcols1)
```

```
finaldf.head()
```

```
finaldf.shape
```

```
X = finaldf.drop(['SALARY'],axis=1)
y= finaldf.SALARY
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=1234)
X_train.shape
```

```
from sklearn.linear_model import LinearRegression
linmodel = LinearRegression()
linmodel.fit(X_train,y_train)
y_pred = linmodel.predict(X_test)
```

```
residuals = pd.DataFrame({'pred':y_pred,'actul':y_test})
residuals.head()
```

```
from sklearn.metrics import r2_score
r2_score(y_pred,y_test)
```
