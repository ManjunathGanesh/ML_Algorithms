```
catcols = df.dtypes[df.dtypes =='object'].index.tolist()
catcols
```

```

def onehotfun(df,catcol):
  onehotdummies = []
  onehotdummies.append(df)
  for cat in catcol:
    catname = pd.get_dummies(df[cat],prefix=cat)
    onehotdummies.append(catname)

  dffinal = pd.concat(onehotdummies,axis=1)
  dffinal.drop(columns=catcol, axis=1, inplace=True)
  return dffinal
```

```
finaldf = onehotfun(df,catcols)
```
