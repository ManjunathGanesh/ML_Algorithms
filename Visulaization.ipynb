```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```
data = [['Brilliant', 51,7.4], ['Genius', 65.5,4.7], ['Winners', 58.2,3.1]]
```

```
df = pd.DataFrame(data, columns = ['School', 'Avg_Height','Std_Height'])
```

```
#sns.displot(x=Avg_Heights, kde = False, color ='red', bins = 30)
#sns.histplot(x =Avg_Heights, y =school)
sns.displot(df, x="Avg_Height",y="Std_Height", hue="School")
```
