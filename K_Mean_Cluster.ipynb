```python
import pandas as pd
import seaborn as sns
```

```python
penguin = sns.load_dataset('penguins')
```

```python
penguin.head()
```

```python
penguin = penguin[~penguin.body_mass_g.isna()]
```

```python
penguin.head()
```

```python
sns.scatterplot(data=penguin,x='flipper_length_mm',y='body_mass_g')
```

```python
from sklearn.cluster import KMeans
```

```python
km= KMeans(n_clusters=4)
```

```python
km.fit(penguin[['flipper_length_mm','body_mass_g']])
```

```python
grps = km.predict(penguin[['flipper_length_mm','body_mass_g']])
```

```python
grps[:10]
```

```python
penguin.loc[:,'cluster']=grps
```

```python
penguin.head()
```

```python
sns.scatterplot(data=penguin,x='flipper_length_mm',y='body_mass_g',hue='cluster')
```

```python
penguin[penguin.cluster==3].species.value_counts()
```

```python
penguin[penguin.cluster==2].species.value_counts()
```

```python
penguin[penguin.cluster==1].species.value_counts()
```
