```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#Importing Penguin Cleaned data set from git hub account

```python
penguin = pd.read_csv("https://raw.githubusercontent.com/Manju410/MLPractice/main/data/Penguin.csv")
penguin.head()
```

# Rows and Columns of penguin data set

```python
penguin.shape
```

# Checking Null values in penguin data set

```python
penguin.isna().sum()
```

# Independent and dependent variables of penguin data set

```python
X = penguin.drop("species",axis=1)
y = penguin.species
```

```python
X.shape,y.shape
```

# KNN Classifier Model

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_classifier = KNeighborsClassifier()
```

```python
knn_classifier
```

```python
param_grid = {"n_neighbors":range(2,31)}
```

# Grid SearchCv to find best parameter from model

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(knn_classifier,param_grid,scoring='accuracy',cv=10)
```

```python
grid_model.fit(X,y)
```

# grid model contain n_neighbors from range 2 to 30 with k fold cv=10

```python
df_grid = pd.DataFrame(grid_model.cv_results_)
df_grid.columns
```

```python
df_grid[['rank_test_score','params','mean_test_score','std_test_score']].sort_values('rank_test_score')
```

```python
grid_model.best_score_
```

```python
grid_model.best_params_
```

```python
grid_model.best_estimator_
```

# From above grid search model we found n_neighbors=10 to fit best KNN clasifier model

```python
knn_model = grid_model.best_estimator_
```

```python
knn_model.score(X,y)
```

```python
y_pred = knn_model.predict(X)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y,y_pred)
```
