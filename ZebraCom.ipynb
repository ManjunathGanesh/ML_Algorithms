#A> What is the probability of finding a random sample of 50 Zebra handsets, having their average battery life lower than 17 hours ?

```
def random_num(low, high, size):
   return [np.random.randint(low, high) for _ in range(size)]
```

```
np.random.randint?
```

```
Zebra_Fifty = np.array(random_num(700, 1400, 50))
```

```
Zebra_Fifty.shape
```

```
sns.histplot(Zebra_Fifty);
plt.axvline(1080, color='r', ls='--', label='Average hours')
plt.legend()
```

```
rng = np.random.default_rng(20)
```

```
Zebra_smpls = np.array([rng.choice(Zebra_Fifty, size=50) for _ in range(10_000)])
```

```
Zebra_smpls.shape
```

```
mean_smpls = Zebra_smpls.mean(axis=1)
```

```
sns.histplot(mean_smpls);
```

```
sampleMean = mean_smpls.mean()
sampleMean
```

```
samplestd = mean_smpls.std()
samplestd
```
