```
def ConvertToInteger(dataframe,column,typ):
  dataframe[column]=pd.to_numeric(dataframe[column],errors='coerce')
  dataframe=dataframe.dropna(subset=[column])
  dataframe.loc[:,column]=dataframe[column].astype(typ)
  return dataframe
```

```
# Using ECDF to understand the spread of data

def ecdf(data):
  n_data=len(data)
  x_data=np.sort(data)
  y_data=np.linspace(1/n_data,1,n_data)
  return x_data, y_data
```

- less than 60% of innings he scored less than 40 runs
- less than 80% of innings he scored less than 100 runs
- every 20% of innings he scored more than 100 runs

```
plt.figure(figsize=(20,6))
x_vals,y_vals=ecdf(vk['Runs'])
plt.plot(x_vals,y_vals, marker='.',ls='none',label='Virat Kholi')

_= plt.xlabel('Runs scored')
_=plt.ylabel('ECDF')
plt.legend()
```

```
plt.figure(figsize=(12,5))
plr_dfs=list()
plr_bins=0
 
for name in players:
  pdf=df[df['Name']==name]
  plr_dfs.append(pdf['Runs'])

  n_pdf=int(np.sqrt(len(pdf['Runs'])))
  plr_bins= n_pdf if n_pdf > plr_bins else plr_bins


_=plt.hist(plr_dfs,bins=plr_bins, label=players)
_=plt.xlabel('Runs scored')
_=plt.ylabel('Frequency')
_=plt.legend()
```

- Dhoni played more no of innings and scored less than 

- Dhoni has scored less than 50 runs in more no of innings than others.
