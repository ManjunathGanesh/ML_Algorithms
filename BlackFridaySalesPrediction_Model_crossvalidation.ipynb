# Checking for null values in dataset

```python
Saledf.isna().sum()
```

```python
def Gender(x):
  if x=='F':return 0
  else: return 1
```

```python
Saledf.Gender = Saledf.Gender.apply(Gender)
```

```python
Saledf.head()
```

# converting categorical values to numbers for City category column

```python
Saledf.City_Category.unique()
```

```python
citydf = pd.get_dummies(Saledf.City_Category, prefix='City')
```

```python
Saledf1 = pd.concat([Saledf, citydf], axis=1)
```

```python
Saledf1.drop(columns=['City_Category'], axis=1, inplace=True)
Saledf1.head()
```

# Extract Train and test dataset

```python
X= Saledf1.drop(columns=['Purchase','Product_ID'],axis=1)
y= Saledf1.Purchase.values.reshape(-1,1)
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(
    X,y,test_size=0.25, random_state=1234
)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

# Cross Validation Technique

```python
from sklearn.model_selection import KFold
```

```python
from sklearn.model_selection import cross_val_score
```

```python
X_new = pd.concat([X_train,X_test])
X_new.head()
```

```python
X_new.shape
```

```python
X_new = X_new.reset_index().drop('index',axis=1)
```

```python
X_new.head()
```

```python
y_new = pd.concat([pd.DataFrame(y_train,columns=['Price']),pd.DataFrame(y_test,columns=['Price'])])
y_new.head()
```

```python
y_new.shape
```

```python
scores = cross_val_score(model,X_new,y_new,
                         cv=KFold(n_splits=4,shuffle=True,random_state=1234),
                         scoring='r2')
```
